```python
import pandas as panda
import numpy as num
import pandas_profiling
```

```python
#Reading csv file using Pandas
data = panda.read_csv("D:\ML\panda_profiling\Meteorite_Landings.csv", parse_dates=['year'],encoding='UTF-8')
data['year']=panda.to_datetime(data['year'], errors='coerce')
# Example: Constant variable
data['source'] = "NASA"

# Example: Boolean variable
data['boolean'] = num.random.choice([True, False], data.shape[0])

# Example: Mixed with base types
data['mixed'] = num.random.choice([1, "A"], data.shape[0])

# Example: Highly correlated variables
data['reclat_city'] = data['reclat'] + num.random.normal(scale=5,size=(len(data)))

# Example: Duplicate observations
duplicates_to_add = panda.DataFrame(data.iloc[0:10])
duplicates_to_add[u'name'] = duplicates_to_add[u'name'] + " copy"

data = data.append(duplicates_to_add, ignore_index=True)
```

```python
pandas_profiling.ProfileReport(data)
```
